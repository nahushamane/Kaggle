```python
# libraries

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# load dataset

titanic_train = pd.read_csv("train.csv")
```

```python
print(titanic_train.shape)
```

```python
print(titanic_train.columns)
```

```python
titanic_train.head()
```

```python
titanic_train["Survived"].value_counts()
```

```python
titanic_train.plot(kind="scatter", x="Age", y="Survived")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.FacetGrid(titanic_train, hue="Survived", size=8).map(plt.scatter, "Sex", "Fare").add_legend()
plt.show()
```

```python
sns.set_style("whitegrid")
sns.pairplot(titanic_train, hue='Survived', size=3)
plt.show()
```
